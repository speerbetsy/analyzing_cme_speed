```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from os import listdir
from io import StringIO
import pickle
import datetime

with open('all_cmes.pkl', 'rb') as f:
    data = pickle.load(f)
#data.to_csv('out.csv', sep=' ')
ht_data=data['HT_DATA']
height_data=ht_data[0]['HEIGHT']
time_data=ht_data[0]['DATE_TIME']

print ("this is time_data: ", time_data[0])
print ("this is height_data: ", height_data)
#new_date= datetime.datetime(1996,1,11,0,25,22)
```

```python
plt.title("Height vs Time")
plt.xlabel('time')
plt.ylabel('Meters')
plt.plot(time_data, height_data,'+')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
diff_time_data=np.diff(time_data)
diff_height_data=np.diff(height_data)
velocity=diff_height_data/diff_time_data.astype(float)

plt.title("Velocity vs Time")
plt.xlabel('Time')
plt.ylabel('Meters/Second')
plt.plot(time_data[:-1], velocity,'+') #one less data point
plt.gcf().autofmt_xdate()
plt.show()
```

```python
diff_diff_time_data=np.diff(diff_time_data)
diff_v_data=np.diff(velocity)
acc=diff_v_data/diff_diff_time_data.astype(float)

plt.title("Acceleration vs Time")
plt.xlabel('Time')
plt.ylabel('Meters/Second^2')
plt.plot(time_data[:-2], acc,'+')
plt.gcf().autofmt_xdate()
plt.show()
```
